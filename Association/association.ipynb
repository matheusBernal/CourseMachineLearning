```python
import pandas as pd
import numpy as np
import plotly.express as px
from apyori import apriori
```

```python
df_market = pd.read_csv('../Bases de dados/mercado.csv',header=None)
df_market
```

```python
transacations = []
for i in range(len(df_market)):
    transacations.append([str(df_market.values[i,j]) for j in range(df_market.shape[1])])
transacations
```

```python
type(transacations)
```

```python
rules = apriori(transacations, min_support=0.3,min_confidence=0.8, min_lift=2)
results = list(rules)
results
```

```python
len(results)
```

```python
results[2]
```

```python
r[1]
```

```python
r[2]
```

```python
A =[]
B = []
support=[]
trust = []
lift = []

for r in results:
    s = r[1]
    results_rule = r[2]
    for result_rule in results_rule:
        a = list(result_rule[0])
        b = list(result_rule[1])
        c = result_rule[2]
        l = result_rule[3]
        A.append(a)
        B.append(b)
        support.append(s)
        trust.append(c)
        lift.append(l)
```

```python
trust
```

```python
lift
```

```python
rules_market = pd.DataFrame({'A':A,'B':B,'Support':support,'Trus':trust,'Lift':lift})
rules_market
```

```python
rules_market.sort_values(by='Lift',ascending=False)
```
