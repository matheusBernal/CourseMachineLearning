```python
from sklearn import datasets
import plotly.express as px
from sklearn.cluster import KMeans,DBSCAN,AgglomerativeClustering
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
x_random,y_random = datasets.make_moons(n_samples=1500,noise=0.09)
```

```python
x_random
```

```python
y_random
```

```python
grafic = px.scatter(x=x_random[:,0],y=x_random[:,1])
grafic
```

```python
kmeans_ = KMeans(n_clusters=2)
routers = kmeans_.fit_predict(x_random)
```

```python
routers
```

```python
grafic = px.scatter(x=x_random[:,0],y=x_random[:,1],color=routers)
grafic
```

```python
hc = AgglomerativeClustering(n_clusters=2,linkage='ward')
routers = hc.fit_predict(x_random)
```

```python
grafic = px.scatter(x=x_random[:,0],y=x_random[:,1],color=routers)
grafic
```

```python
dbsanc = DBSCAN(eps=0.10)
routers = dbsanc.fit_predict(x_random)
routers
```

```python
grafic = px.scatter(x=x_random[:,0],y=x_random[:,1],color=routers)
grafic
```
