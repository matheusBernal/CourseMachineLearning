```python
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
with open('x_credit_risk.plk','rb') as f:
    x_credit_risk,y_credit_risk = pickle.load(f)
```

```python
x_credit_risk = np.delete(x_credit_risk,[2,7,11], axis=0)
```

```python
x_credit_risk
```

```python
y_credit_risk = np.delete(y_credit_risk,[2,7,11], axis=0)
y_credit_risk
```

```python
credit_risk_logistic_regression = LogisticRegression(random_state=1)
credit_risk_logistic_regression.fit(x_credit_risk,y_credit_risk)
```

```python
credit_risk_logistic_regression.intercept_
```

```python
credit_risk_logistic_regression.coef_
```

```python
predict_risk_credit = credit_risk_logistic_regression.predict([[0,0,1,2],[2,0,0,0]])
predict_risk_credit
```

```python
with open('x_credit.plk','rb') as g:
    x_credit_test,y_credit_test,x_credit_train,y_credit_train = pickle.load(g)
```

```python
x_credit_test
```

```python
y_credit_test
```

```python
credit_logistic_regression = LogisticRegression(random_state=1)
credit_logistic_regression.fit(x_credit_train,y_credit_train)
```

```python
credit_logistic_regression.intercept_
```

```python
credit_logistic_regression.coef_
```

```python
predict_credit = credit_logistic_regression.predict(x_credit_test)
```

```python
accuracy_credit = accuracy_score(y_credit_test,predict_credit)
accuracy_credit
```

```python
print(classification_report(y_credit_test,predict_credit))
```
