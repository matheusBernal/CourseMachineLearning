```python
import Orange
```

```python
df_base_credit = Orange.data.Table('..\Bases de dados\credit_data_regras.csv')
```

```python
df_base_credit.domain
```

```python
import Orange.classification


majority = Orange.classification.MajorityLearner()
```

```python
import Orange.evaluation


predictions = Orange.evaluation.testing.TestOnTestData(df_base_credit,df_base_credit,[majority])
```

```python
Orange.evaluation.CA(predictions)
```

```python
for register in df_base_credit:
    print(register.get_class())
```
